```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
#przykładowe dane
powierzchnia = np.array([30,40,50,60,70,80,90,100,120]).reshape(-1,1)
cena = np.array([190000,245000,295000,367000,547000,612000,742900,830500,1110000])
```

```python
#model regresji liniowwj
model = LinearRegression()
model.fit(powierzchnia,cena)
```

```python
print(f"wspóczynnik nachylenia (a): {model.coef_[0]}")
```

```python
print(f"wyraz wolny (b): {model.intercept_}")
```

```python
print(f"R^2 (miara dopasowania): {model.score(powierzchnia,cena)}")
```

```python
#prognoza
pow_75 = np.array([[75]])
prognoza = model.predict(pow_75)
print(f"prognozowana cena dla 75m2: {prognoza[0]}")
```

```python
#wykres
plt.scatter(powierzchnia,cena,label='Dane')
plt.plot(powierzchnia,model.predict(powierzchnia),color='red',label='Regresja liniowa')
plt.scatter(75,prognoza,color='green',label='Prognoza (75m2)')
plt.xlabel('Powierzchnia [m2]')
plt.ylabel('Cena [PLN]')
plt.title('Regresja liniowa: Cena mieszkania vs Powierzchnia')
plt.legend()
plt.show()
```
